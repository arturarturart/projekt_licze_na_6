```python
import numpy as np
import matplotlib.pyplot as plt

def draw_chessboard():
    # Tworzenie szachownicy jako dwuwymiarowej tablicy numpy
    chessboard = np.zeros((8, 8))

    # Wypełnienie szachownicy kolorami na zmianę
    for i in range(8):
        for j in range(8):
            if (i + j) % 2 == 0:
                chessboard[i, j] = 1

    # Rysowanie szachownicy
    plt.imshow(chessboard, cmap='gray')
    plt.axis('off')
    plt.title('Szachownica')

def draw_colored_text(text):
    colors = ['red', 'blue', 'green', 'purple', 'orange', 'pink', 'brown', 'cyan', 'magenta']

    fig, ax = plt.subplots(figsize=(12, 6))

    # Rysowanie liter
    x = 0.1
    for char in text:
        color = np.random.choice(colors)
        ax.text(x, 0.5, char, fontsize=60, weight='bold', color=color, va='center', ha='center')
        x += 0.1  # Odstęp między literami

    # Ukrywanie osi
    ax.axis('off')

    plt.show()

# Wywołanie funkcji
draw_colored_text("Lubię Szachy")
draw_chessboard()
```
