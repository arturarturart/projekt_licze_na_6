```python
import numpy as np
import matplotlib.pyplot as plt

def draw_heart(ax):
    # Tworzenie siatki punktów
    x = np.linspace(-2, 2, 200)
    y = np.linspace(-2, 2, 200)
    X, Y = np.meshgrid(x, y)

    # Równanie serca
    F = (X**2 + Y**2 - 1)**3 - X**2 * Y**3

    # Tworzenie wykresu
    ax.contour(X, Y, F, [0], colors='red')

    # Ustawienia wykresu
    ax.set_aspect('equal')
    ax.set_xlim(-2, 2)
    ax.set_ylim(-2, 2)
    ax.axis('off')

def draw_text_diagonally(ax, text):
    # Rysowanie ukośnych liter
    for i, char in enumerate(text):
        ax.text(0, -i, char, fontsize=20, color='red', ha='left', va='center')

    # Ustawienia osi
    ax.set_xlim(-1, 1)
    ax.set_ylim(-len(text), 1)
    ax.axis('off')

# Tworzenie figury z dwoma osiami
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 8), gridspec_kw={'width_ratios': [1, 1]})

# Rysowanie ukośnego tekstu na pierwszej osi
draw_text_diagonally(ax1, "BOKS JEST MOIM HOBBY")

# Rysowanie serca na drugiej osi
draw_heart(ax2)

# Wyświetlanie wykresu
plt.tight_layout()
plt.show()
```
